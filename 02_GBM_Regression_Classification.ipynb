## **GBM (Classification):    **

```python
#   package 
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import matplotlib

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import r2_score
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
!git clone https://github.com/yun-seon/DX-Training-LGES-0614.git
```

```python
'''
 / dataset :  19  1   (Categorical feature: State, Area_code, Int.l.Plan, VMail.Plan)
: 'Churn.' / True(1)  , False(0)    
'''

#  
dt2 = pd.read_csv("/content/DX-Training-LGES-0614/Data_GBM_cls_churn.csv")
print('dataset shape:', dt2.shape)

#  
dt2.head()
```

```python
#  (type) 
dt2.info()
```

```python
#  

# column name  "." "_" 
dt2.columns = dt2.columns.str.replace(".","_")

# data Area.Code     , category  
'''
- category:     or    (ex.  : 0, 1)
- object:         (ex.  : , , ,  )
'''
dt2.Area_Code = dt2.Area_Code.astype("category")

print(dt2['Churn_'].value_counts())

unsatisfied_cnt = dt2[dt2['Churn_'] == "True." ].Churn_.count()
total_cnt = dt2.Churn_.count()

print('\n  {0:.3f}'.format((unsatisfied_cnt / total_cnt))) # True: 
```

```python
#   (type) 
dt2.info()
```

```python
# (Xs) (Y) 
X_features = dt2.iloc[:, :-1]
y_labels = dt2.iloc[:, -1]

#      ('True.') 1,  0   binary class {1, 0} 
y_labels = np.where(y_labels == "True.", 1, 0)
y_labels = pd.Series(y_labels)

print('type of labels:{0}'.format(np.unique(y_labels)))
```

```python
# object  category 
for col in X_features.columns:
    if X_features[col].dtypes == "O":
        X_features[col] = X_features[col].astype('category')

#  (type) 
X_features.info()
```

```python
# GBM categorical feature  one-hot encoding 
def onehot(data, feature):

    return pd.concat([data, pd.get_dummies(data[feature], prefix=feature)], axis=1).drop([feature], axis=1)

#  4 categorical feature  one-hot encoding    X_feature 
cate_list = ['State', 'Area_Code', 'Int_l_Plan', 'VMail_Plan']

for i in range(len(cate_list)):

    X_features = onehot(X_features, cate_list[i])
```

```python
#  feature 
X_features.info()
```

```python
#     
#   : 0.8,   : 0.2
train_X, test_X, train_y, test_y = train_test_split(X_features, y_labels, test_size=0.2, random_state=0)
train_cnt = train_y.count()
test_cnt = test_y.count()

print('  Shape:{0},   Shape:{1}'.format(train_X.shape , test_X.shape))

print('\n     ')
print(train_y.value_counts()/train_cnt)

print('\n     ')
print(test_y.value_counts()/test_cnt)
```

```python
# GBM  package 
from sklearn.ensemble import GradientBoostingClassifier

# GBM    .   
start_time = time.time()

#    train data fitting () 
clf = GradientBoostingClassifier(loss             = 'deviance',
                                 criterion        = 'squared_error', #    
                                 n_estimators     = 100,
                                 min_samples_leaf = 5,
                                 max_depth        = 3,
                                 learning_rate    = 1,
                                 random_state     = 0)
clf.fit(train_X , train_y)

# GBM     
gb_time = time.time() - start_time
print("GBM  : {0:.2f}  ".format(gb_time))
```

image.png

```python
#   GBM     
pred = clf.predict(test_X)

#       , () 
accuracy = accuracy_score(test_y, pred)
print('GBM : {0:.4f}'.format(accuracy))
```

```python
# precision/recall/f1-score/accuracy 
print(classification_report(test_y, pred))
```

```python
#       , AUC 
roc_score = roc_auc_score(test_y, clf.predict_proba(test_X)[:,1], average='macro')
print('ROC AUC: {0:.4f}'.format(roc_score))
```

auc-roc-curve.png

```python
#    , GBM      

# feature importance  / feature importance   feature     
feature_importance = clf.feature_importances_

# feature_importance   
sorted_idx = np.argsort(feature_importance)

#   plot X    
pos = np.arange(sorted_idx.shape[0]) + .5

fig = plt.figure(figsize = (10, 10))
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, np.array(X_features.columns)[sorted_idx])
plt.title('Feature Importance')
fig.tight_layout()
plt.show()
```

```python
#   ( )
for i, col in enumerate(dt2.iloc[:, :-1].columns):
    print(f'{col}  : {clf.feature_importances_[i]}')

print("\n", clf.get_params()) # GradientBoostingClassifier    
```

> GridSearch   

```python
#  hyper parameter   GridSearchCV 
from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators':[100, 300, 500],
    'learning_rate' : [ 0.05, 0.1]
}

gb_grid_cv = GridSearchCV(clf , param_grid=params , cv=2 ,verbose=1, n_jobs=-1)
gb_grid_cv.fit(train_X , train_y)
print('  :\n', gb_grid_cv.best_params_)
print('  : {0:.4f}'.format(gb_grid_cv.best_score_))
```

```python
# GridSearchCV    estimator  
gb_pred = gb_grid_cv.best_estimator_.predict(test_X)
gb_accuracy = accuracy_score(test_y, gb_pred)
print('GBM : {0:.4f}'.format(gb_accuracy))

gb_grid_cv_roc_score = roc_auc_score(test_y, gb_grid_cv.predict_proba(test_X)[:,1],average='macro')
print('GBM ROC AUC: {0:.4f}'.format(gb_grid_cv_roc_score))
```
